```python
def mark_map(m,key="price_avg",minv=0,maxv=10000):
    for e in m.mapeles:
        try:
            m.remove_layer(e)
        except:
            pass
    m.mapeles=[]
    for n,c in cities.items():
        if key in c:
            try:
                if c[key]<=maxv and c[key]>=minv:
                    mark = CircleMarker(location=[c["lat"],c["lon"]],opacity=0,radius=10,
                                        fill_color = "red",
                                        fill_opacity=0.4,
                                        weight=1,
                                  #color=colors[int(cl)].hex
                                 )
                    html_value = "<div><h4>"+n+"</h4><table>"+"".join(["<tr><td>"+str(key)+"</td><td>"+str(val)+"</td></tr>" for key,val in c.items()])+"<table></div>"
                    mark.popup = HTML(html_value,placeholder='',description='',)
                    
                    m += mark
                    m.mapeles.append(mark)
                    
            except Exception as e:
                print(e)
                pass
        
        #    cl=(c.price["mean"]-minp) * (len(colors) - 1) / (maxp-minp)
              
```
